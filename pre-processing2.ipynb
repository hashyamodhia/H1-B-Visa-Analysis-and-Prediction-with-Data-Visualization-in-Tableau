#### Converting data types of int and float columns into their respective least required datatype.

```python
def reduce_memory_usage(df):
    for column in preprocessed_data1.columns:
        column_dtype = df[column].dtype
        if column_dtype != object:
            column_min = df[column].min() #To store the minimum value in that column 
            column_max = df[column].max() # to store the maximum value in that column

            # now if a column is not object it can either be integer or float
            if str(column_dtype)[:3] == 'int':
                # check in range of which dataype does min and maximm values lie and set the data type accordingly of that column.
                # minimum value of the column should be greater than or equal to lower bound of that datatype and and maximum value of that column 
                # should be lesser than or equal to upperbound of that data type.
                # np.iinfo gives integer info of given datatype say np.int8
                if column_min >= np.iinfo(np.int8).min and column_max <= np.iinfo(np.int8).max:
                    df[column] = df[column].astype(np.int8)
                elif column_min >= np.iinfo(np.int16).min and column_max <= np.iinfo(np.int16).max:
                    df[column] = df[column].astype(np.int16)
                elif column_min >= np.iinfo(np.int32).min and column_max <= np.iinfo(np.int32).max:
                    df[column] = df[column].astype(np.int32)
                else:
                    df[column] = df[column].astype(np.int64)

            # Now if a column is not object or int then it'll be float
            # Now here we'll use finfo to get float info of given datatype like float16
            else:
                if column_min >= np.finfo(np.float16).min and column_max <= np.finfo(np.float16).max:
                    df[column] = df[column].astype(np.float16)
                elif column_min >= np.finfo(np.float32).min and column_max <= np.finfo(np.float32).max:
                    df[column] = df[column].astype(np.float32)
                else: 
                    df[column] = df[column].astype(np.float64)
    return df
    
```

```python
preprocessed_data1 = reduce_memory_usage(preprocessed_data1)
preprocessed_data1.info()
```
