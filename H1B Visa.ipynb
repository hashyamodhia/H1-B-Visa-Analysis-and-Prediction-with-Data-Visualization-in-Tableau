```python
from sklearn.utils import resample

df_down_sampled_certified = resample(df_certified, n_samples=210000, random_state=42)
df_down_sampled_certified
```

```python
df_remaining = preprocessed_data3[preprocessed_data3["CASE_STATUS"] != 0]
df_remaining
```

```python
balanced_dataset = pd.concat([df_remaining, df_down_sampled_certified])
balanced_dataset
```

```python
balanced_dataset = balanced_dataset.sample(frac = 1, random_state = 42)
balanced_dataset = balanced_dataset.reset_index(drop = True)  # Drop = True discards the old index
balanced_dataset
```

```python
balanced_dataset["CASE_STATUS"].value_counts()
```

```python
balanced_dataset.info()
```

```python
x = balanced_dataset.drop('CASE_STATUS', axis = 1)
y = balanced_dataset['CASE_STATUS']
x
```

```python
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
# Splitting the data
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=42)
```

```python
# Initialize the CatBoost model
catboost_model = CatBoostClassifier()
catboost_model.fit(x_train, y_train)

# Evaluate the model on the training set
y_pred_train_catboost = catboost_model.predict(x_train)
catboost_training_accuracy = accuracy_score(y_train, y_pred_train_catboost)

# Evaluate the model on the test set
y_pred_test_catboost = catboost_model.predict(x_test)
catboost_accuracy = accuracy_score(y_test, y_pred_test_catboost)
catboost_precision = precision_score(y_test, y_pred_test_catboost, average='weighted')
catboost_recall = recall_score(y_test, y_pred_test_catboost, average='weighted')
catboost_f1 = f1_score(y_test, y_pred_test_catboost, average='weighted')
catboost_cm = confusion_matrix(y_test, y_pred_test_catboost)

# Store the performance metrics for CatBoost
catboost_performance_chart = [
    ("CatBoost", catboost_training_accuracy, catboost_accuracy, catboost_precision, catboost_recall, catboost_f1, catboost_cm)
]
```

```python
# Initialize the XGBoost model
xgboost_model = XGBClassifier()
xgboost_model.fit(x_train, y_train)

# Evaluate the model on the training set
y_pred_train_xgboost = xgboost_model.predict(x_train)
xgboost_training_accuracy = accuracy_score(y_train, y_pred_train_xgboost)

# Evaluate the model on the test set
y_pred_test_xgboost = xgboost_model.predict(x_test)
xgboost_accuracy = accuracy_score(y_test, y_pred_test_xgboost)
xgboost_precision = precision_score(y_test, y_pred_test_xgboost, average='weighted')
xgboost_recall = recall_score(y_test, y_pred_test_xgboost, average='weighted')
xgboost_f1 = f1_score(y_test, y_pred_test_xgboost, average='weighted')
xgboost_cm = confusion_matrix(y_test, y_pred_test_xgboost)

# Store the performance metrics for XGBoost
xgboost_performance_chart = [
    ("XGBoost", xgboost_training_accuracy, xgboost_accuracy, xgboost_precision, xgboost_recall, xgboost_f1, xgboost_cm)
]
```

```python
performance1_df = pd.DataFrame(catboost_performance_chart, columns=['Model Name', 'Training Accuracy', 'Accuracy', 'Precision', 'Recall', 'F1', 'Confusion Matrix'])
performance1_df
```
