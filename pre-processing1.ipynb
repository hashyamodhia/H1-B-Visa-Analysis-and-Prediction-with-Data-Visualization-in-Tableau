### Converting prevailing wages to hours and years and their unit of pays.

```python
final_dataset["PREVAILING_WAGE"].isna().sum()
```

```python
final_dataset["PW_UNIT_OF_PAY"].isna().sum()
```

```python
final_dataset["FULL_TIME_POSITION"].value_counts()
```

```python
final_dataset["FULL_TIME_POSITION"].isna().sum()
```

```python
import pandas as pd
import numpy as np

def convert_wage(row):
    if pd.isna(row['PREVAILING_WAGE']):
        return np.nan  # Keep NA values as they are
    
    wage = row['PREVAILING_WAGE']
    unit = row['PW_UNIT_OF_PAY']
    full_time = row['FULL_TIME_POSITION']
    
    if unit == 'Hour':
        if full_time == 'Y':
            return wage * 40 * 48  # Full-time Hourly to yearly considerin 40 hour work week and 48 woking weeks in an year
        else:
            return wage * 20 * 48  # Part-time Hourly to yearly considering 20 hour work week and 48 working weeks in an year.
    elif unit == 'Week':
        return wage * 48  # Considering 48 hours work week in an year
        
    elif unit == 'Bi-Weekly':
        return wage * 24  # Considering 48 work weeks in an year
        
    elif unit == 'Month':
        return wage * 12  # Monthly to yearly
    elif unit == 'Year':
        return wage  # Yearly remains the same
    else:
        return np.nan  # In case of an unexpected unit

def convert_unit_to_year(row):
    unit = row['PW_UNIT_OF_PAY']
    
    if unit == 'Hour':
        return 'Year'  # Keep Hour as is
    elif unit == 'Week':
        return 'Year'  # Convert Week to Year
    elif unit == 'Bi-Weekly':
        return 'Year'  # Convert Bi-Weekly to Year
    elif unit == 'Month':
        return 'Year'  # Convert Month to Year
    elif unit == 'Year':
        return 'Year'  # Already in Year
    else:
        return np.nan  # In case of an unexpected unit

# Assuming final_dataset is your DataFrame
final_dataset['PREVAILING_WAGE'] = final_dataset.apply(convert_wage, axis=1)
final_dataset['PW_UNIT_OF_PAY'] = final_dataset.apply(convert_unit_to_year, axis=1)
```
